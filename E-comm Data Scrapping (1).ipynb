## DATA SCRAPING

```python
#importing librarires 
from bs4 import BeautifulSoup
import requests, openpyxl
import pandas as pd
```

```python
source = requests.get('https://www.flipkart.com/search?q=laptop&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&as-pos=1&as-type=RECENT&suggestionId=laptop%7CLaptops&requestId=4c86673d-7e0c-496d-9422-bf4b154a9d7c&as-backfill=on&p%5B%5D=facets.price_range.from%3DMin&p%5B%5D=facets.price_range.to%3DMax')
```

```python
excel = openpyxl.Workbook()
sheet = excel.active
sheet.title = 'Laptop REVIEW'
sheet.append(['Name', 'Processor', 'Rating', 'Price', 'Warranty'])
```

```python
soup = BeautifulSoup(source.text, 'html.parser')
print(soup.prettify())
```

```python
products= [] #List to store name of the product
prices= [] #List to store price of the product
ratings = []
processors = []
warr = []

try:
    results = soup.find('div', class_="_1YokD2 _3Mn1Gg").find_all('div', class_="_1AtVbE col-12-12")

    for result in results:
        name = result.find('div', class_="_4rR01T").text.split('-')[0]          
        price = result.find('div', class_="_30jeq3 _1_WHN1").text
        try:
            rating =  result.find('div',  class_="_3LWZlK").text
        except: rating = "unrated"
        specs = result.find('ul',  class_="_1xgFaf").text.split(' ')
        processor =  result.find('li',  class_="rgWa7D").text
        warranty = []
        if specs[-1]=="Warranty":
            warranty = "Yes"
        else:
            warranty = "No"
            
        products.append(name)
        prices.append(price)
        ratings.append(rating)
        processors.append(processor)
        warr.append(warranty)
        
        df = pd.DataFrame({'Name': products, 'Rating': ratings, 'Processor': processors, 'Price': prices, 'Warranty': warr})
        df.to_excel("Laptop.xlsx", index=False, encoding='utf-8')

except Exception as e:
    print(e)
```
